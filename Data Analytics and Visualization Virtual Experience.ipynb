### Import Modules & Datasets

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
Content=pd.read_csv('D:\Datasets\Forage Project Data\Content.csv')
Reaction=pd.read_csv('D:\Datasets\Forage Project Data\Reactions.csv')
ReactionTypes=pd.read_csv('D:\Datasets\Forage Project Data\ReactionTypes.csv')
Profile=pd.read_csv('D:\Datasets\Forage Project Data\Profile.csv')
Session=pd.read_csv('D:\Datasets\Forage Project Data\Session.csv')
```

```python
Content.head()
```

```python
Content.drop('Unnamed: 0',axis=1,inplace=True)
Reaction.drop(['Unnamed: 0','User ID'],axis=1,inplace=True)
ReactionTypes.drop('Unnamed: 0',axis=1,inplace=True)
Profile.drop('Unnamed: 0',axis=1,inplace=True)
Session.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
Content.head(2)
```

```python
data=Content.merge(Profile,how='left',on='User ID')
data=data.merge(Session,how='left',on='User ID')
data=data.merge(Reaction,how='left',on='Content ID')
data=data.rename(columns={'Type_x':'Content Type','Type_y':'Type'})
data=data.merge(ReactionTypes,how='left',on='Type')
data=data.rename(columns={'Type':'Reaction Type'})
```
